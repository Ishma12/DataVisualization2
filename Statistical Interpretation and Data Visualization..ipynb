6.2 Data Transformation: [4]
Write code to transform variables according to the following instructions:
1. ”School”, ”sex”, ”address”, ”schoolsup”, ”famsup”, ”activites”, ”nursery”, ”internet”, and ”romantic”
into binary: 0 or 1 (create new columns without overwriting the existing ones).

```python
#inserting the replaced value (0,1)of all variables into new column respectively.
df['school1'] = df[['school']].replace(['GP','MS'], [0,1])
df['sex1'] = df[['sex']].replace(['F','M'], [0,1])
df['address1'] = df[['address']].replace(['U','R'], [0,1])
df['schoolsup1'] = df[['schoolsup']].replace(['no','yes'], [0,1])
df['famsup1'] = df[['famsup']].replace(['no','yes'], [0,1])
df['activities1'] = df[['activities']].replace(['no','yes'], [0,1])
df['nursery1'] = df[['nursery']].replace(['no','yes'], [0,1])
df['internet1'] = df[['internet']].replace(['no','yes'], [0,1])
df['romantic1'] = df[['romantic']].replace(['no','yes'], [0,1])

df
```

2. ”Medu”, ”Fedu”, ”reason”, ”guardian”, ”studytime”, ”freetime”, and ”health” into ordinal numbers based
on the number cases in the data set (create news columns without overwriting the existing ones).

```python
#here none in medu & fedu is unique value.
medu = {'none': 0, 'primary education (4th grade)': 1, '5th to 9th grade':2, 'secondary education':3 ,'higher education': 4}
df['Medu1'] = [medu[item] for item in df.Medu]

fedu = {'none': 0, 'primary education (4th grade)': 1, '5th to 9th grade':2, 'secondary education':3 ,'higher education': 4}
df['Fedu1'] = [fedu[item] for item in df.Fedu]
#r for reason
r = {'course': 1, 'home': 2, 'reputation': 3, 'other': 4}
df['reason1'] = [r[item] for item in df.reason]
#g for guardian
g = {'mother': 1, 'father': 2, 'other': 3}
df['guardian1'] = [g[item] for item in df.guardian]
#st for studytime
st = {'2 to 5 hours': 3, '5 to 10 hours': 5, '<2 hours': 1, '>10 hours': 6}
df['studytime1'] = [st[item] for item in df.studytime]
#ft for freetime
ft = {'very low': 1, 'low': 2, 'medium': 3, 'high': 4, 'very high': 5}
df['freetime1'] = [ft[item] for item in df.freetime]
#h for health
h = {'very bad': 1, 'bad': 2, 'good': 3, 'very good': 4, 'excellent': 5}
df['health1'] = [h[item] for item in df.health]

df
```

3. Convert column ”age” to interval datatype. i.e. Create a new column name category age whose values
should be based on the frequency in the column ”age”, You can create categorical data with following
interval.
(a) interval1: [15-17]
(b) interval2: [18-20]
(c) interval3: [21-all]

```python
print(df[['age']].to_string()) #printing the age of 395  students with its index 0-394

#using for loop and if conditions 
#first make a three empty lst, declaring the intervals and append the items in lst if it appears in that interval.
lst1 = []
lst2 = []
lst3 = []
#1 value is given to the students whose age appear interval between 14-17, 2 for 17-20 and 3 for 20-23
interval1 = pd.Interval(14, 17)
interval2 = pd.Interval(17, 20)
interval3 = pd.Interval(20, 23)

for items in df['age']:
  if items in interval1:
    lst1.append(items)

for items in df['age']:
  if items in interval2:
    lst2.append(items)

for items in df['age']:
  if items in interval3:
    lst3.append(items)

#adding another column category age and insert the values that we append in the lists. 
df['category age'] = df[['age']].replace([lst1, lst2, lst3], [1, 2, 3])
df
```

Create a new column name passed (yes or no) whose values should be based on the values present in the

```python
x = (df['G3'])
high = x [x >= 8] #storing number more than and equals to 8 in variable high
low = x [x < 8]   #storing number less than 8 in variable low

df['Passed'] = df[['G3']].replace([high, low], ['yes' , 'no']) #replacing the value of high with yes and low with no
df
```
